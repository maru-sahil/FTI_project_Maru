```python
import pandas as pd

# Load the dataset
url = "https://raw.githubusercontent.com/maru-sahil/FTI_project_Maru/refs/heads/main/zip_water_quality.csv"
df = pd.read_csv(url)

# Drop duplicates based on City, keeping only the first occurrence
df_unique = df.drop_duplicates(subset=["city"], keep="first")

# Show the cleaned DataFrame
print(df_unique)

# Optionally, save to a new CSV
df_unique.to_csv("zip_water_quality_unique.csv", index=False)


```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Load your dataset (update the path to where your file is saved)
df = pd.read_csv("https://raw.githubusercontent.com/maru-sahil/FTI_project_Maru/refs/heads/main/zip_water_quality.csv")

# Scatter plot: population vs. total violations
plt.figure(figsize=(8,6))
plt.scatter(df["population"], df["total_violations"], alpha=0.6, color="teal")

# Labels and title
plt.xlabel("Population Served")
plt.ylabel("Total Violations")
plt.title("Do Bigger Water Systems Have More Violations?")

# Optional: log scale helps if population varies a lot
plt.xscale("log")

plt.show()
```

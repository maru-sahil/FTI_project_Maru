```python
import pandas as pd

# Load the dataset
url = "https://raw.githubusercontent.com/maru-sahil/FTI_project_Maru/refs/heads/main/zip_water_quality.csv"
df = pd.read_csv(url)

# Drop duplicates based on City, keeping only the first occurrence
df_unique = df.drop_duplicates(subset=["city"], keep="first")

# Show the cleaned DataFrame
print(df_unique)

# Optionally, save to a new CSV
df_unique.to_csv("zip_water_quality_unique.csv", index=False)


```
